```python
NC['UnemploymentRate'] = (NC['UnemployedCiviliansE']/NC['CiviliansInLaborceE']) * 100
```

```python
components =  ['BelowPovertyE', 'WhiteAloneE',
       'BlackorAfricanAmericanaloneE', 'AmericanIndianandAlaskaNativealoneE',
       'AsianaloneE', 'NativeHawaiianandOtherPacificIslanderaloneE',
       'SomeotherracealoneE', 'TwoormoreracesE',
       'TwoormoreracesTworacesincludingSomeotherraceE',
       'TworacesexcludingSomeotherraceandthreeormoreracesE',
       'MedianHouseholdIncomeE', 'UnemploymentRate', 'pct_occupied', 'rent_per_month',
       'tpoverty', 'tminority', 'spending_per_month',
       'BlackorAfricanAmericanaloneERescaled', 'GrossRentScaled',
       'IncomeInequalityIndex', 'OpportunityIndex', 'GEOID11',
       'Minorityshare']
```

```python
components
```

```python
PC = pd.DataFrame(NC,columns = components)
PC.dtypes
```

```python
PC = PC.loc[:,('BelowPovertyE',
 'WhiteAloneE',
 'BlackorAfricanAmericanaloneE',
 'AmericanIndianandAlaskaNativealoneE')].apply(np.float64)
```

```python
PC.dtypes
```

```python
scaler = MinMaxScaler()
scaler.fit(PC)
scaled = scaler.transform(PC)
```

```python
scaled
```

```python
PC = PCA(n_components =1)
PC.fit(scaled)
```

```python
xPC = PC.transform(scaled)
xPC.shape
```

```python
print(PC.explained_variance_ratio_)
```

```python
NC['PCA'] = xPC
```
