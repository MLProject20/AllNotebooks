There are two ways to answer this question. First, you can use a search cursor, which will allow you to to extract the coordinates of parcel centroid and each bus stop. The distance between the two can be determined using the Euclidean metric, see below for an example.

```python
### Read shapefile for busstops, parcels and line 

route9Bustops = featureclasses[13]
route9parcels = featureclasses[15]
route9line = featureclasses[14]

route9Bustops, route9parcels, route9line
```

```python
### Create a list called radius and ridership
### use range to get specficied distance
## Use select by location to select parcels that are with the different distances of the bustops
## Use Get count to get the number of ridership from each select by location run
## append the radius and ridership to thier respective list



radius = []
ridership  = []

for row in range(100,401,10):
    management =  arcpy.management.SelectLayerByLocation(route9parcels, "WITHIN_A_DISTANCE", route9Bustops, str(row) + " Meters",
                                       "NEW_SELECTION", "NOT_INVERT")
    count = int(arcpy.GetCount_management(management)[0]) 
    radius.append(row)
    ridership.append(count)
    
    print("The radius " + str(row) + "m" + " " + " ridership is " + str(count))
```

```python
### Loop over the origin and destination for bustops and censusblockgroupcentroids respectively
### calculate the euclidean distance betweeen the origin and the destination 
### Create a point for each origin/destination and add them to an array
### use the array to create a polyline
### Use search cursor to inser the polyline inside the polyline feature class created above 

## Set threshold distance 

Threshold = 2500


for row in range(0,len(Xori),1):
    for col in range(0,len(XDes),1):
        distance = sqrt((XDes[col] - Xori[row])**2 + (YDes[col]-Yori[row])**2)
        print(distance)
       
        if distance < Threshold:
            array = arcpy.Array([arcpy.Point(Xori[int(row)], Yori[int(row)]),
                                 arcpy.Point(XDes[int(col)], YDes[int(col)])])
            polyline = arcpy.Polyline(array)
            
            with arcpy.da.InsertCursor(FC,["SHAPE@", "Population"]) as cursor:
                cursor.insertRow([polyline,Pop[row]])
            del cursor
```

16. Using the digital elevation model for Crater Lake, prepare a script that will generate maps where you will automatically change the number of classes. You should choose the quantile classification, and generate maps from 2 classes to 25 classes. Every time you change the number of classes, you will change the title accordingly.

Note that you can access the classification properties using the symbology property of the raster. We have talked about this in class (April 19), but you can find more information in the link below. https://pro.arcgis.com/en/pro-app/latest/arcpy/mapping/symbology-class.htm

```python
## specify aprx for the maps which is called RasterLake.aprx
## List the maps in the ArcGIS pro project(aprx)
### List the layers
### set a range for the class 
### specify the breakcount
### specify the classification method
### list the layout inside the ArcGIS Project
### Check the text dynamically
### Export layer as pdf 

print("Script Start: " + strftime("%Y-%m-%d %H:%M:%S"))



p = arcpy.mp.ArcGISProject("RasterLake.aprx")
m = p.listMaps('Map')[0]
a = m.listLayers()
Lake = a[0]


for row in range(2,26,1):
    
    sym = Lake.symbology
    sym.colorizer.classificationField = "Value"
    sym.colorizer.breakCount = row
    sym.colorizer.colorRamp = p.listColorRamps('Purples (Continuous)')[0]
    sym.colorizer.classificationMethod = "Quantile"
    Lake.symbology = sym
    print(sym.colorizer.breakCount)
    l = p.listLayouts()[0]
    l.listElements('TEXT_ELEMENT')[0].text = "Lake Crater " + str(row) + " classes"
    l.exportToPDF("LakeCrater_classes"+str(row)+".pdf")
    
    
print("Script End: " + strftime("%Y-%m-%d %H:%M:%S"))
```
